search name table, get back unique list of tax_ids
query taxonomy table and return name from there

```python
# model.py
from sqlobject import *
import os.path
dbfile = 'taxa.db3'

conn_str = os.path.abspath(dbfile)
conn_str = 'sqlite:'+ conn_str
sqlhub.processConnection = connectionForURI(conn_str)
class Taxonomy(SQLObject):
    class sqlmeta:
        idName = "tax_id"
        fromDatabase = True
    names = MultipleJoin('Name', joinColumn="tax_id")

class Name(SQLObject):
    class sqlmeta:
        fromDatabase = True
    taxa = ForeignKey('Taxonomy', dbName="tax_id")
```

```python
query = 'Homo sapien'

condition = Name.q.name.contains(query)

id_set = set()
for n in Name.select(condition):
    id_set.add(n.taxID)

for i in id_set:
    print(Taxonomy.get(i).scientificName, 'tax')
```
